```python
import pandas as pd
df= pd.read_csv("pharmacy_messy (1).csv")
print("shape",df.shape)
df.head()
```

```python
df.columns= df.columns.str.lower().str.strip().str.replace(" ", " ")
print(df.columns.tolist())
```

```python
df=df.rename(columns={"receipt id":"receipt_id", "customer type":"customer_type", "drug category":"drug_category", "total amount":"total_amount", "payment method":"payment_method", "unit price":"unit_price", "pharmacist name":"pharmacist_name"})
print("Final column names")
print(df.columns.tolist())
```

```python
print("Type Before:",df["unit_price"].dtype)
print(df["unit_price"].head())
```

```python
df["unit_price"] = df["unit_price"].astype(str)
df["unit_price"]= df["unit_price"].str.replace("GHS", " ").str.strip()
df["unit_price"]= pd.to_numeric(df["unit_price"], errors="coerce")
print("Type after:", df["unit_price"].dtype)
print(df["unit_price"].head())
```

```python
print("Type before", df['date'].dtype)
print(df['date'].head())
```

```python

df["month"]=df["date"].dt.month
df['day']=df['date'].dt.day
df[["date", "month", "day"]].head()
```

```python
df.dtypes
```


- What was the type before and after for Unit Price?

Before: Object, After: Float64

- What was the type before and after for Date?


Before: Object,   After: Datetime64


```python
print(df.isnull().sum())
```

```python
df["total_amount"]= df["total_amount"].fillna(df["total_amount"].mean())

df["rating"]= df["rating"].fillna(df["rating"].mean())

print("missing in total_amount:", df["total_amount"].isnull().sum())
print("missing in rating:", df["rating"].isnull().sum())
```

```python
df["pharmacist"]= df["pharmacist"].fillna("Unknown")


print("missing in pharmacist:", df["pharmacist"].isnull().sum())
```

```python
print("Remaining missing values:")
print(df.isnull().sum())
```

```python
print("Rows before dropping", len(df) )

df=df.dropna()

print("Rows after dropping", len(df))
```

```python
print(df.isnull().sum())

print("\nFinal shape:", df.shape)
```
